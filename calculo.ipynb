```python
import pandas as pd
import os
```

```python
caminho = os.getcwd()

arquivo = pd.read_excel('nova.xlsx')
indice = pd.read_excel('indice.xlsx')
```

```python
display(arquivo)
```

```python
lista_datas = (arquivo.iloc[0][7]).split(',')

lista_valores = (arquivo.iloc[0][4]).split(',')

for i, data in enumerate(lista_datas):
    teste = pd.to_datetime(data)
    
    #pega a data da lista de datas dos meses da divida
    data_mes_div = f'{teste.month}-{teste.year}'
    
    #pega a linha indice do mes da divida
    ind_mes_div = indice[indice['data abreviada'] == data_mes_div]
    
    #pega o valor do indice do mes da divida --referindo a linha completa anterior
    valor_ind_div = float(ind_mes_div['indice'].values[0])
    
    #mes e dia no mesmo formato ja para pesquisa exp '05-2024'
    hoje_formatado = f'{pd.Timestamp.today().month}-{pd.Timestamp.today().year}'
    
    #verifica se o mes atual tem registro de indice no banco de dados
    try:
        indice_mes_atual = indice[indice['data abreviada'] == hoje_formatado]
        valor_indice_mes_atual = float(indice_mes_atual['indice'].values[0])
    
    #caso o mes atual nao tenha um registro de indice ele ira puxar do ultimo mes, caso seja o primeiro mes do ano ele ira puxar do ano anterior
    except:
        mes,ano = hoje_formatado.split('-')
        if int(mes)-1 ==0:
            mes = 12
            ano = int(ano)-1
            nova_data =f'{mes}-{ano}'
        else:
            mes = int(mes) - 1
            nova_data =f'{mes}-{ano}'
        indice_mes_atual = indice[indice['data abreviada'] == nova_data]
        valor_indice_mes_atual = float(indice_mes_atual['indice'].values[0])
    
    indice_att = 94.458606/valor_ind_div
    valor_div_corre =  int(lista_valores[i]) * indice_att
    
    print(arquivo.iloc[i][1])
    print(f'total de parcelas: {len(lista_datas)}')
    print(f'referente a fatura: {i+1}')
    print(f'valor divida sem atualização: {lista_valores[i]}')
    print(f'valor da divida atualizado: {valor_div_corre:.02f}')
    print(f'valor indice atualizado: {float(indice_att):.02f}')
    print(f'valor indice mes da divida: {valor_ind_div}')
    print(f'indice atual: {float(94.458606):.02f}')
    print(f'data mes divida: {data_mes_div}')
    print('\n ')
    

```
